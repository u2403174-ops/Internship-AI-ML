```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/content/health.csv')
df.head()
```

```python
missing= df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 6))
ax = missing.plot(
    kind='bar',
    color='skyblue',
    edgecolor='black',
    width=0.8
)
plt.title('Missing Values per Column', fontsize=16, fontweight='bold')
plt.ylabel('Missing Values', fontsize=12)
plt.xlabel('Column', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
for index, value in enumerate(missing):
    plt.text(
        index,
        value + max(missing)*0.01,
        str(value),
        ha='center',
        va='bottom',
        fontsize=9,
        fontweight='bold'
    )

plt.tight_layout()
plt.show()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
categorical_cols = [
    'Gender',
    'Blood Type',
    'Medical Condition',
    'Doctor',
    'Hospital',
    'Insurance Provider',
    'Admission Type',
    'Medication',
    'Test Results'
]

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
test_result_counts = df['Test Results'].value_counts()

print("Test Result Counts:\n")
print(test_result_counts)
```

```python
sns.countplot(x='Test Results', data=df, hue='Test Results', palette='Set2', legend=False)

plt.title('Count of Test Results')
plt.xlabel('Test Results')
plt.ylabel('Patient Count')

plt.show()
```

```python
sns.countplot(x='Gender', data=df, hue='Gender', palette='pastel', legend=False)

plt.title('Patient Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Patient Count')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.kdeplot(
    data=df,
    x='Age',
    fill=True,
    color='purple'
)

plt.title('KDE Plot of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Density')

plt.show()
```

```python
test_result_counts = df['Test Results'].value_counts()

plt.figure(figsize=(7,7))

plt.pie(
    test_result_counts,
    labels=test_result_counts.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Test Results Distribution')

plt.show()
```
